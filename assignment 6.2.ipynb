```
import sklearn
from sklearn.datasets import fetch_openml
import pandas as pd
import matplotlib.pyplot as plt


titanic = fetch_openml(name='titanic', version=1, as_frame=True, parser='auto')
df_titanic = titanic.data
df_titanic.columns = titanic.feature_names
print(df_titanic.head())
```

```
import seaborn as sns
sns.boxplot(df_titanic['age'])
```

```
import seaborn as sns
import matplotlib.pyplot as plt


def removal_box_plot(df, column, threshold):
    sns.boxplot(df[column])
    plt.title(f'Original Box Plot of {column}')
    plt.show()

    removed_outliers = df[df[column] <= threshold]

    sns.boxplot(removed_outliers[column])
    plt.title(f'Box Plot without Outliers of {column}')
    plt.show()
    return removed_outliers


threshold_value = 0.12

no_outliers = removal_box_plot(df_titanic, 'age', threshold_value)
```
